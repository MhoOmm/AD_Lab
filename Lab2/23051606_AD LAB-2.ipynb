0.4.1. Find Sum and Average of List in Python

   Sample Input:-\
      a = [10, 20, 30, 40, 50]\
   Output:\
    Sum of the list: 150 \
    Average of the list: 30.0

```python
def action(a):
    sum1 = sum(a)
    avg = sum1 / len(a)  
    return sum1,avg 

a = [10, 20, 30, 40, 50]
sum1,avg = action(a)
print("Sum of the list: ",sum1)
print("Average of the list:",avg)
```

0.4.4. Get Last N Elements from Given List

   For Example:
   
   Input : test_list = [4, 5, 2, 6, 7, 8, 10]

   Output : [2, 6, 7, 8, 10]

```python
def getlast(test,n):
    size = len(test)
    if(n>size):
        print("enter a smaller number")
    i = size - n
    result = []
    for it in range(i,size):
        result.append(test[i])
        i+=1
        
    return result
         
test_list = [4,5,2,6,7,8,10]
n = int(input("enter the number of elements required"))
print(getlast(test_list,n))
```

0.4.11. Python program to capitalize the first and last character of each word in a string

For Example:  

Input: s = "hello world"

Output: HellO WorlD

```python
def caplast(s):
    words = s.split()
    res= []
    for w in words:
        if len(w) == 1:
            res.append(w.upper())
        else:
            new = w[0].upper() + w[1:-1] + w[-1].upper()
            res.append(new)
    return " ".join(res)


s = input("Enter the string")
print("Entered string:",s)
print(caplast(s))
```

0.4.12. Maximum and Minimum element in a Set in Python

For Example:  

Input: s1 = {4, 12, 10, 9, 4, 13}

```python
def find(s):
    minimum = min(s)
    maximum = max(s)
    return minimum, maximum

s1 = {4, 12, 10, 9, 4, 13}
mn, mx = find(s1)
print("Entered set:",s1)
print("Minimum element:", mn)
print("Maximum element:", mx)
```

0.4.15. Sort Python Dictionary by Key or Value 

For Example:  

Input: d = {'ravi': 10, 'rajnish': 9, 'sanjeev': 15}

Output:{'rajnish': 9, 'ravi': 10, 'sanjeev': 15}

```python
def sort(d):
    key = list(d.keys())
    key.sort()
    sd = {}
    for k in key:
        sd[k] = d[k]
    return sd
d = {'ravi': 10, 'rajnish': 9, 'sanjeev': 15}
print("original dictionary:",d)
snew = sort(d)
print("sorted dictionary: ",end ='')
print(snew)
```

0.4.17. Counting the Frequencies in a List Using Dictionary in Python

For Example:  

Input: ['apple', 'banana', 'apple', 'orange', 'banana', 'banana'] 

Output: {'apple': 2, 'banana': 3, 'orange': 1}

```python
def frequency(a):
    freq = {}
    for item in a:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1 
    return freq

a = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
print("Entered list:",a)
print(frequency(a))
```
