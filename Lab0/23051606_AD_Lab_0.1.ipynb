```python
sec = int(input("enter time in seconds: "))

if sec < 60:
    print("0 hour 0 minute", sec, "seconds")

else:
    h = sec // 3600
    min = (sec % 3600) // 60
    sec = sec % 60
    print(h, "hours", min, "mins", sec, "seconds")
```
