```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Smaller Dataset

```python
df = pd.read_csv('house_price_dataset_small.csv')
df.head()
```

Importing the libraries

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,root_mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Size (sq ft)']]
y = df.iloc[:,-1]
X.shape
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
X_train.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred  = lr.predict(X_test)
print("Mean_squared_error:",mean_squared_error(y_test,y_pred))
print("Mean_absolute_error:",mean_absolute_error(y_test,y_pred))
print("Root_mean_squared_error:",root_mean_squared_error(y_test,y_pred))
print("R2_Score:",r2_score(y_test,y_pred))
small = []
small.append(mean_squared_error(y_test,y_pred))
small.append(mean_absolute_error(y_test,y_pred))
small.append(root_mean_squared_error(y_test,y_pred))
small.append(r2_score(y_test,y_pred))
```

Larger Dataset

```python
dflarge = pd.read_csv('house_price_dataset_large.csv')
dflarge.head()
```

```python
X = dflarge[['Size (sq ft)']]
y = dflarge['Price ($)']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
X_train.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred  = lr.predict(X_test)
print("Mean_squared_error:",mean_squared_error(y_test,y_pred))
print("Mean_absolute_error:",mean_absolute_error(y_test,y_pred))
print("Root_mean_squared_error:",root_mean_squared_error(y_test,y_pred))
print("R2_Score:",r2_score(y_test,y_pred))

large = []
large.append(mean_squared_error(y_test,y_pred))
large.append(mean_absolute_error(y_test,y_pred))
large.append(root_mean_squared_error(y_test,y_pred))
large.append(r2_score(y_test,y_pred))
```

```python
result = pd.DataFrame(
    {
        'Small Dataset': small,
        'Large Dataset': large
    },
    index=['MSE', 'MAE','RMSE', 'R2']
)

result
```
