3. Sales Data GroupBy Analysis

Read sales data from the sales_data_10.csv file and:

a) Calculate total sales per product

b) Calculate average sales per product

Sample Input: sales_data_10.csv

The content of csv file is

Product,Sales

Pen,100

Pencil,150

Pen,120

Eraser,80

Pencil,170

Pen,130

Marker,200

Eraser,90

Marker,180

Pencil,160

Sample Output:

Product Total_Sales Average_Sales

Eraser 170 85.00

Marker 380 190.00

Pen 350 116.67

Pencil 480 160.00

```python
df = pd.read_csv('sales_data_10.csv')
grp=df.groupby('Product')

sum = grp['Sales'].sum()
avg = grp['Sales'].mean()

result = pd.DataFrame({
    'Total_Sales': sum,
    'Average_Sales': avg
}).reset_index()

print(result)
```

Read student marks from the missing_marks_10.csv file and:

```python
df = pd.read_csv('missing_marks_10.csv')
df['IsMissing'] = df['Marks'].isna()
df['Replaced Marks'] = np.where(df['IsMissing'],np.mean(df['Marks']),df['Marks'])
print(df)
```
