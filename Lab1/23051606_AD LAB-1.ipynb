0.3.5 WAP to check whether a number is prime or not

```python
def prime(n):
    n = int(input("Enter a number: "))
    if n <= 1:
        print("Not a prime number")
    else:
        is_prime = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
    return is_prime

num  = int(input("Enter the number"))
if prime(num):
        print("Prime number")
else:
        print("Not a prime number")
```

0.3.10 WAP to find out the prime factors of a number entered through keyboard (distinct).

```python
def primefactors(n):
    factors = set()

    while n % 2 == 0:
        factors.add(2)
        n //= 2

    i = 3
    while i * i <= n:
        while n % i == 0:
            factors.add(i)
            n //= i
        i += 2
        
    if n > 2:
        factors.add(n)

    return factors

num = int(input("Enter a number: "))
result = primefactors(num)

print("Distinct prime factors:", result)
```

0.3.20. WAP to form reverse pyramid of numbers for a given number. Ex. for number 4

```python
# 1 2 3 4 3 2 1
#   1 2 3 2 1
#     1 2 1
#       1
def reversepyramid(n):
    for i in range(n, 0, -1):
        print(" " * (n - i) * 2, end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        for j in range(i - 1, 0, -1):
            print(j, end=" ")
        print()

num = int(input("Enter a number: "))
reversepyramid(num)

```

0.3.23 WAP to convert a decimal number into its equivalent number with base b. Decimal number and b are the user input.

```python
def db(num, base):
    if num == 0:
        return "0"
    digits = "0123456789ABCDEF"
    result = ""
    while num > 0:
        rem = num % base
        result = digits[rem] + result
        num = num // base
    
    return result

decimal = int(input("enter a decimal number: "))
base = int(input("enter the base to convert to: "))
print("Number in base :",base,"is:",db(decimal, base))
```

0.3.24 WAP to convert a number with base b into its equivalent decimal number. Number with base b & b are the user input.

```python
def bd(num, base):
    num = num.upper()
    digits = "0123456789ABCDEF"
    power = 0
    decimal = 0
    
    for i in range(len(num) - 1, -1, -1):
        digit = num[i]
        value = digits.index(digit)
        decimal += value * (base ** power)
        power += 1

    return decimal

num = input("Enter number with base b: ")
b = int(input("Enter the base: "))

print("Number and base entered:",num,",",b)

print("Equivalent decimal number =", bd(num, b))
```

0.3.25 WAP to convert a binary number to its equivalent octal & hexa-decimal number system.

```python
def bd(binary):
    power = 0
    decimal = 0
    for i in range(len(binary) - 1, -1, -1):
        bit = int(binary[i])
        decimal += bit * (2 ** power)
        power += 1
    return decimal


def do(n):
    result = ""
    if n == 0:
        return "0"
    while n > 0:
        result = str(n % 8) + result
        n //= 8
    return result


def dh(n):
    digits = "0123456789ABCDEF"
    result = ""
    if n == 0:
        return "0"
    while n > 0:
        result = digits[n % 16] + result
        n //= 16
    return result


binary = input("Enter a binary number: ")

decimal = bd(binary)
octal = do(decimal)
hexadecimal = dh(decimal)

print("Binary number entered:",binary)
print("Octal equivalent =", octal)
print("Hexadecimal equivalent =", hexadecimal)
```
